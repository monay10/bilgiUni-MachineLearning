```python
import pandas as pd # dataframe kullanımı için
import numpy as np #mathematiksel işlemler için
import matplotlib.pyplot as plt # çizdirme kütüphanesi
import seaborn as sns # gelişmiş çizdirme kütüphanesi 
%matplotlib inline
```

```python
df = pd.read_csv('train.csv')
```

```python
df.drop(['Customer ID','Network type subscription in Month 1','Network type subscription in Month 2','Most Loved Competitor network in in Month 1','Most Loved Competitor network in in Month 2'],axis =1, inplace=True)# anlamsız sütunları sildim
```

```python
newdf = df.iloc[:1399]
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split # train ve test datalarını ayırmak için kullanılan kütüphane
```

```python
X = newdf.drop('Churn Status',axis=1)
y = newdf['Churn Status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier(random_state=101)
```

```python
dtree.fit(X_train,y_train) # % 80 lik train kısmının  X ve Y 'sini birbirine öğrettim.Supervised learning'in temelidir.
```

```python
predictions = dtree.predict(X_test) # tahminler üretildi elimdeki Y_test ile kıyaslayacağım
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
import math
```

```python
mse = mean_squared_error(y_test,predictions)# burada root mean square error u buldum.
mae = mean_absolute_error(y_test,predictions)#  burada mean absolute error u buldum
```

```python
mae
```

```python
rmse = math.sqrt(mse)
```

```python
rmse # burada root mean square error u buldum.
```

```python
from sklearn.metrics import classification_report,confusion_matrix,mean_absolute_error,mean_squared_error
```

```python
print(classification_report(y_test,predictions)) # prediction ile Y_Test arasında matrix kurdum
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=44)
```

```python
from sklearn.ensemble import RandomForestClassifier
error_rate = []
for x in range(50,500,50):
    rfc = RandomForestClassifier(n_estimators=x, random_state=44)
    rfc.fit(X_train, y_train)
    rfc_pred = rfc.predict(X_test)
    error_rate.append(1- np.mean(rfc_pred != y_test))
    
```

```python
rfc = RandomForestClassifier(n_estimators=400, random_state=44)
rfc.fit(X_train, y_train)
rfc_pred = rfc.predict(X_test)
error_rate.append(1- np.mean(rfc_pred != y_test))
```

```python
print(confusion_matrix(y_test,rfc_pred))
```

```python
print(classification_report(y_test,rfc_pred))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(50,500,50),error_rate,color='blue', linestyle = 'dashed',marker = 'o',markerfacecolor = 'red', markersize = 10)
plt.title('Doğruluk  Oranı vs N Estimator Sayısı')
plt.xlabel('N')
plt.ylabel( 'Doğruluk Oranı')
```

```python
print(classification_report(y_test,rfc_pred))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=44)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error_rate = []

for i in range(1,50):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_knn = knn.predict(X_test)
    error_rate.append(1- np.mean(pred_knn != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,50),error_rate,color='blue', linestyle = 'dashed',marker = 'o',markerfacecolor = 'red', markersize = 10)
plt.title('Doğruluk Oranı vs K referans sayısı')
plt.xlabel('K Referansı')
plt.ylabel( 'Doğruluk Oranı')
```

```python
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train, y_train)
pred_knn = knn.predict(X_test)
error_rate.append(1- np.mean(pred_knn != y_test))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,pred_knn))
```

```python
error_rate = []
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
from sklearn import svm
for i in range(1,20):
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    clf = svm.SVC(kernel='linear',random_state=44, C = i)
    clf.fit(X_train,y_train)   
    y_pred = clf.predict(X_test)    
    error_rate.append(1- np.mean(y_pred != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,20),error_rate,color='blue', linestyle = 'dashed',marker = 'o',markerfacecolor = 'red', markersize = 10)
plt.title('Doğruluk Oranı vs C değeri')
plt.xlabel('C değeri')
plt.ylabel( 'Doğruluk Oranı')
```
